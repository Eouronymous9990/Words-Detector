```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

# Set the path to the Tesseract OCR executable
pytesseract.pytesseract.tesseract_cmd = r"E:\vsc\tesseract.exe"
img = cv2.imread(r"C:\Users\zbook 17 g3\Desktop\New folder (4)\TEXTO.png")
hi,w,_=img.shape
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_thickness = 2
font_color = (0, 0, 255)

rects=pytesseract.image_to_data(img,lang="eng")

print(rects)

for a,b in enumerate(rects.splitlines()):
    if a!=0:
        b=b.split()
        print(b)
        if len(b)==12:
            coords = [int(b[i]) for i in [6, 7, 8, 9]]
            x, y, w, h = coords
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),6)
            cv2.putText(img,b[11],(x,y+80),font, font_scale, font_color, font_thickness)
#cv2.imshow("image",img) 
#cv2.waitKey(0)
#cv2.destroyAllWindows()   
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Text")  
plt.axis("off")  
```
